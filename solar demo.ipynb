```python
import numpy as np
import matplotlib.pyplot as plt

power0 = 100
years = 10
t = np.arange(years + 1)

const_deg = 0.02
rad_400 = 0.01
rad_700 = 0.025

def degrade(p0, const, rad):
    p = [p0]
    for _ in range(years):
        p.append(p[-1] * (1 - const - rad))
    return np.array(p)

p400 = degrade(power0, const_deg, rad_400)
p700 = degrade(power0, const_deg, rad_700)

plt.plot(t, p400 / power0 * 100, label="400 km")
plt.plot(t, p700 / power0 * 100, label="700 km")

plt.xlabel("Years")
plt.ylabel("Power (%)")
plt.legend()
plt.grid()
plt.show()
```
